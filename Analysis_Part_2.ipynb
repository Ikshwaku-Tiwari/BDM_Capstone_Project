```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Final_Data_BDM_colab.csv")
```

```python
df['Sales'] = df['Sales'].replace({'â‚¹': '', ',': ''}, regex=True)
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
```

```python
fabric_data = {
    'Fabric': ['Satin', 'Linen', 'Poplin Cotton', 'Tusser Cotton', 'Synthetic'],
    'Price Range': ['360-430', '400-440', '430-450', '460-480', '360-400'],
    'Min Price': [360, 400, 430, 460, 360],
    'Max Price': [430, 440, 450, 480, 400],
}

workforce_data = {
    'Job': ['Ironing', 'Cutting', 'Stitching', 'Finishing and Packaging'],
    'No of Employees': [1, 2, 5, 2],
}

fabric_df = pd.DataFrame(fabric_data)
workforce_df = pd.DataFrame(workforce_data)
```

```python
plt.figure(figsize=(10, 6))
for i, row in fabric_df.iterrows():
    plt.plot([row['Fabric'], row['Fabric']], [row['Min Price'], row['Max Price']], marker='o')
plt.xticks(rotation=45)
plt.title('Price Range per Dress Type')
plt.xlabel('Material Used')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(workforce_df['No of Employees'], labels=workforce_df['Job'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Blues"))
plt.title('Employee Distribution by Job Type')
plt.axis('equal')
plt.show()
```

```python
import folium
from folium.plugins import HeatMap
from geopy.distance import geodesic

pune_coords = (18.5204, 73.8567)
cost_per_km = 10

m = folium.Map(location=pune_coords, zoom_start=6)


for index, row in df.iterrows():
    other_coords = (row["Latitude"], row["Longitude"])
    distance_km = geodesic(pune_coords, other_coords).km
    transport_cost = distance_km * cost_per_km

    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=(
            f"<b>City:</b> {row['City/Town']}<br>"
            f"<b>Distance from Pune:</b> {distance_km:.2f} km<br>"
            f"<b>Transport Cost:</b> â‚¹{transport_cost:.2f}"
        ),
        icon=folium.Icon(color="blue")
    ).add_to(m)

    folium.PolyLine(
        locations=[pune_coords, [row["Latitude"], row["Longitude"]]],
        color="green",
        weight=2,
        opacity=0.7
    ).add_to(m)

m.save("transport_cost_map.html")
m
```

```python
from geopy.distance import geodesic
pune_coords = (18.5204, 73.8567)
cost_per_km = 10

unique_cities = df.groupby("City/Town").first().reset_index()
results = []
for index, row in unique_cities.iterrows():
    other_coords = (row["Latitude"], row["Longitude"])
    distance_km = geodesic(pune_coords, other_coords).km
    transport_cost = distance_km * cost_per_km

    results.append({
        "City/Town": row["City/Town"],
        "Distance (km)": round(distance_km, 2),
        "Transport Cost (â‚¹)": round(transport_cost, 2)
    })

for result in results:
    print(f"City/Town: {result['City/Town']}, Distance: {result['Distance (km)']} km, Transport Cost: â‚¹{result['Transport Cost (â‚¹)']}")
```
